```python
def EncodeTarget(Target):
    dict={
        'Enrolled' :0,
        'Dropout' :1,
        'Graduate':2
    }
    return dict[Target]
df['Target']=df['Target'].apply(EncodeTarget)
```

```python
def nomalized_data(df,col_names) :
    nomalized_df=df.copy()
    for col in col_names:
        mean=nomalized_df[col].mean()
        std=nomalized_df[col].std()
        nomalized_df[col]=(nomalized_df[col]-mean)/std
    return nomalized_df
```

```python
col_to_nomalize=['Previous qualification (grade)',
                 'Inflation rate',
                 'GDP',
                 'Unemployment rate',
                 'Course']
nomalized_df=nomalized_data(df,col_to_nomalize)
```

```python
nomalized_df.head()
```

```python
X=nomalized_df.iloc[:,:-1]
y=nomalized_df.iloc[:,-1]
```

```python
encode_y=np.zeros((y.shape[0],3))
for i in range(encode_y.shape[0]):
    encode_y[i,y[i]]=1
```

```python
encode_y
```

```python
class SoftmaxRegression():
    def __init__(self,epoch,lr):
        self.lr=lr
        self.epoch=epoch
        self.acc_array=[]
        self.losses=[]

    def softmax(self, z):
        return np.exp(z)/np.sum(np.exp(z),axis=0)
    
    def loss_function(self, y, y_pred):
        return -((y*np.log(y_pred+1e-6)).sum(axis=-1)).mean()
    
    def accuracy(self,y,y_pred):
        return (1-np.abs(y-y_pred)).mean()
    
    def fit(self, X, y):
        n=X.shape[0]
        d=X.shape[1]
        num_class=y.shape[-1]
        self.theta=np.zeros((d,num_class))
        with tqdm(range(self.epoch)) as pb :
            for i in pb:
                pb.set_description(f"epoch :{i+1}")
                y_pred=self.predict(X)
                gradient=(1/n*np.matmul(X.T,(y_pred-y)))
                self.theta=self.theta-self.lr*gradient

                #evaluating
                y_pred=self.predict(X)
                loss=self.loss_function(y,y_pred)
                acc=self.accuracy(y.argmax(axis=-1),y_pred.argmax(axis=-1))
                pb.set_postfix({
                    "loss":loss,
                    "accuracy": acc
                })
                self.losses.append(loss)
                self.acc_array.append(acc)

    def predict(self, X):
        z=np.matmul(X,self.theta)
        y_pred=self.softmax(z)
        return y_pred
```

```python
model=SoftmaxRegression(500,0.5)
model.fit(X,encode_y)
y_pred=model.predict(X)
```

```python
y_pred=model.predict(X)
for i, j in zip(encode_y,y_pred):
    print(i ,j)
```
