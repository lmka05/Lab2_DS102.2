# Map target sang số 

```python
def EncodeTarget(Target):
    dict={
        'Enrolled' :0,
        'Dropout' :1,
        'Graduate':2
    }
    return dict[Target]
df['Target']=df['Target'].apply(EncodeTarget)
```

# Nomalize

```python
def nomalized_data(df,col_names) :
    nomalized_df=df.copy()
    for col in col_names:
        mean=nomalized_df[col].mean()
        std=nomalized_df[col].std()
        nomalized_df[col]=(nomalized_df[col]-mean)/std
    return nomalized_df
```

```python
col_to_nomalize=['Previous qualification (grade)',
                 'Inflation rate',
                 'GDP',
                 'Unemployment rate',
                 'Course']
nomalized_df=nomalized_data(df,col_to_nomalize)
```

# Chia tập train, test

```python
X=nomalized_df.iloc[:,:-1]
y=nomalized_df.iloc[:,-1]
```

```python
Matrix = X.to_numpy()
size = int(0.8*Matrix.shape[0])
X_train, X_test = Matrix[:size], Matrix[size:]
y_train, y_test = y[:size], y[size:]
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# Xây dựng Softmax Regression

```python
import numpy as np
import matplotlib.pyplot as plt

class SoftmaxRegression:
    def __init__(self, learning_rate=0.03, n_iters=1000, verbose=True):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.verbose = verbose
        self.weights = None
        self.bias = None
        self.loss_history = []

    def _softmax(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))  # tránh overflow như lỗi trước đó (tràn số)
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

    # Hàm onehot để chuyển đổi nhãn thành dạng one-hot encoding
    def _one_hot(self, y, num_classes):
        one_hot = np.zeros((y.size, num_classes))
        one_hot[np.arange(y.size), y] = 1
        return one_hot

    def _cross_entropy_loss(self, y_true, y_pred):
        # Thêm epsilon tránh log(0)
        eps = 1e-15
        y_pred = np.clip(y_pred, eps, 1 - eps)
        return -np.mean(np.sum(y_true * np.log(y_pred), axis=1))

    def fit(self, X, y):
        n_samples, n_features = X.shape
        n_classes = len(np.unique(y))

        self.weights = np.zeros((n_features, n_classes))
        self.bias = np.zeros((1, n_classes))

        y_one_hot = self._one_hot(y, n_classes)

        for i in range(self.n_iters):
            logits = np.dot(X, self.weights) + self.bias
            probs = self._softmax(logits)

            # Tính loss và lưu lại
            loss = self._cross_entropy_loss(y_one_hot, probs)
            self.loss_history.append(loss)

            
            dw = (1 / n_samples) * np.dot(X.T, (probs - y_one_hot))
            db = (1 / n_samples) * np.sum(probs - y_one_hot, axis=0, keepdims=True)

            # Update
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            if self.verbose and (i % 100 == 0 or i == self.n_iters - 1):
                print(f"Epoch {i+1}/{self.n_iters}, Loss: {loss:.4f}")

    def predict(self, X):
        logits = np.dot(X, self.weights) + self.bias
        probs = self._softmax(logits)
        return np.argmax(probs, axis=1)

    def predict_proba(self, X):
        logits = np.dot(X, self.weights) + self.bias
        return self._softmax(logits)
```

```python
model=SoftmaxRegression(n_iters=2000, learning_rate=0.01)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(classification_report(y_test, y_pred))
```
