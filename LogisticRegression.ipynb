Số hóa cột Target

```python
def ConvertTarget(Target : str)-> int :
    dict={"Graduate": 1,
         "Dropout": 0,
         "Enrolled":0}
    return dict[Target]
df['Target']=df['Target'].apply(ConvertTarget)
```

Tiền xử lý dữ liệu

```python
def nomalized_data(df,col_names) :
    nomalized_df=df.copy()
    for col in col_names:
        mean=nomalized_df[col].mean()
        std=nomalized_df[col].std()
        nomalized_df[col]=(nomalized_df[col]-mean)/std
    return nomalized_df
    
```

```python
col_to_nomalize=['Previous qualification (grade)',
                 'Inflation rate',
                 'GDP',
                 'Unemployment rate']
nomalized_df=nomalized_data(df,col_to_nomalize)
```

```python
Matrix=df.to_numpy()
Train_size=int(Matrix.shape[0]*0.8)
X_train=Matrix[0:Train_size,:-1]
X_test=Matrix[Train_size:,:-1]
y_train=Matrix[0:Train_size,-1]
y_test=Matrix[Train_size:,-1]
```

Xây dựng Logistic Regression

```python
from tqdm import tqdm
class Logistic_Regression :
    def __init__(self,epoch,lr):
        self.lr=lr
        self.epoch=epoch
        self.acc_array=[]
        self.losses=[]

    def sigmoid(self, z):
        return 1/(1+np.exp(-z))
    
    def loss_function(self, y, y_pred):
        return -(y*np.log(y_pred+(1e-6))+(1-y)*np.log(1-y_pred+(1e-6))).mean()
    
    def accuracy(self,y,y_pred):
        return (1-np.abs(y-y_pred)).mean()
    
    def fit(self, X, y):
        n=X.shape[0]
        d=X.shape[1]
        self.theta=np.zeros((d,1))
        with tqdm(range(self.epoch)) as pb :
            for i in pb:
                pb.set_description(f"epoch :{i+1}")
                y_pred=self.predict(X)
                gradient=(1/n*np.matmul(X.T,(y_pred-y)))
                self.theta=self.theta-self.lr*gradient

                #evaluating
                y_pred=self.predict(X)
                loss=self.loss_function(y,y_pred)
                acc=self.accuracy(y,y_pred)
                pb.set_postfix({
                    "loss":loss,
                    "accuracy": acc
                })
                self.losses.append(loss)
                self.acc_array.append(acc)
                
    def predict(self, X):
        z=np.matmul(X,self.theta)
        y_pred=self.sigmoid(z)
        return y_pred
```

```python
model=Logistic_Regression(500,0.5)
model.fit(X_train,y_train)
```

```python
y_hat=model.predict(X_test)
for i,j in zip(y_test.tolist(),y_hat.tolist()):
    print(f"y:{i}  \t  y_hat :{j}")
```

```python
plt.figure(figsize=(12,8))
plt.plot(model.acc_array,model.losses)
plt.show()
```
